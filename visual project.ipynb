```python
#LET'S WRITE A FUNCTION

def featurize(path):
    img_path = path
    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    features = np.ndarray.flatten(x)
    return features

def featurize2(path):
    img_path = path
    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    return x
```

```python
# compositional metaphors: choose the source, the target and the later the target's *modifier*
source_term = "sunset"#"snow", "creationpillar"
target_term = "sky"#"carpet", "pillar"
featurize_function = featurize #featurize

indir = 'images'

source = []
for root, dirs, filenames in os.walk(indir):
    for f in filenames:
        if source_term in f:
            img_path = 'images/'+str(f)
            feats = featurize_function(img_path)
            source.append(feats)
            
print("number of source instances: ",len(source))

# target main term
target = []
for root, dirs, filenames in os.walk(indir):
    for f in filenames:
        if target_term in f:
            img_path = 'images/'+str(f)
            feats = featurize_function(img_path)
            target.append(feats)
            
print("number of target instances: ",len(target))

#category unity
dataSetI = sum(source[:-1])
dataSetII = source[-1]
result = 1 - spatial.distance.cosine(dataSetI, dataSetII)
print("similarity of a new source: ",result)

dataSetI = sum(target[:-1])
dataSetII = target[-1]
result = 1 - spatial.distance.cosine(dataSetI, dataSetII)
print("similarity of a new target: ",result)


#source-target similarity
dataSetI = sum(source)
dataSetII = sum(target)
result = 1 - spatial.distance.cosine(dataSetI, dataSetII)
print("source -> target similarity: ",result)
```

```python
#adding the target's modifier
target_modifier = "fire"#"white", "comet"
# target modifier
modifier =[]
for root, dirs, filenames in os.walk(indir):
    for f in filenames:
        if target_modifier in f:
            img_path = 'images/'+str(f)
            feats = featurize_function(img_path)
            modifier.append(feats)
            
print("number of target modifier instances: ",len(modifier))

#source to target+modifier
dataSetI = sum(source)
dataSetII = sum(target)+sum(modifier)
result = 1 - spatial.distance.cosine(dataSetI, dataSetII)
print("source -> target+modifier similarity: ",result)
```

```python
def compo_met(source_term, target_term, target_modifier):  
        
    featurize_function = featurize #featurize
    
    indir = 'images'

    source = []
    for root, dirs, filenames in os.walk(indir):
        for f in filenames:
            if source_term in f:
                img_path = 'images/'+str(f)
                feats = featurize_function(img_path)
                source.append(feats)

    print("number of source instances: ",len(source))

    # target main term
    target = []
    for root, dirs, filenames in os.walk(indir):
        for f in filenames:
            if target_term in f:
                img_path = 'images/'+str(f)
                feats = featurize_function(img_path)
                target.append(feats)

    print("number of target instances: ",len(target))

    #category unity
    dataSetI = sum(source[:-1])
    dataSetII = source[-1]
    result = 1 - spatial.distance.cosine(dataSetI, dataSetII)
    print("similarity of a new source: ",result)

    dataSetI = sum(target[:-1])
    dataSetII = target[-1]
    result = 1 - spatial.distance.cosine(dataSetI, dataSetII)
    print("similarity of a new target: ",result)


    #source-target similarity
    dataSetI = sum(source)
    dataSetII = sum(target)
    result1 = 1 - spatial.distance.cosine(dataSetI, dataSetII)
    print("source -> target similarity: ",result1)
    
    
    # target modifier
    modifier =[]
    for root, dirs, filenames in os.walk(indir):
        for f in filenames:
            if target_modifier in f:
                img_path = 'images/'+str(f)
                feats = featurize_function(img_path)
                modifier.append(feats)

    print("number of target modifier instances: ",len(modifier))

    #source to target+modifier
    dataSetI = sum(source)
    dataSetII = sum(target)+sum(modifier)
    result2 = 1 - spatial.distance.cosine(dataSetI, dataSetII)
    print("source -> target+modifier similarity: ",result2)
    
    return result1,result2
```
